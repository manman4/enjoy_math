```python
from sympy import *
init_printing()
```

```python
a,b,c,d,e,f,g,h = symbols('a b c d e f g h')
```

```python
def show(eq):
  display(eq)
  print("を展開すると")
  display(expand(eq))
```

## 相加相乗（偶数次）

```python
# 4変数相加相乗
eq = (a**2-b**2)**2 + (c**2-d**2)**2 + 2*(a*b-c*d)**2
show(eq)
```

```python
# 6変数相加相乗
eq = (a**2+b**2+c**2)*((b**2-c**2)**2+(c**2-a**2)**2+(a**2-b**2)**2)/2 \
   + (d**2+e**2+f**2)*((e**2-f**2)**2+(f**2-d**2)**2+(d**2-e**2)**2)/2 \
   + 3*(a*b*c - d*e*f)**2
show(eq)
```

```python
# 8変数相加相乗
eq = (a**4-b**4)**2+(c**4-d**4)**2+2*(a**2*b**2-c**2*d**2)**2 \
   + (e**4-f**4)**2+(g**4-h**4)**2+2*(e**2*f**2-g**2*h**2)**2 \
   + 4*(a*b*c*d-e*f*g*h)**2
show(eq)
```
